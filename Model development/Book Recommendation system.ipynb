# Collaborative based filtering Recommender system

```python
book_with_rating = rating_data.merge(book_data,on="ISBN")
```

```python
book_with_rating.head()
```

For making this type of recommender we need to have some criteria, so the criteria we are selecting is we are including user who have given atleast 100 reviews and we will include a book on which atleast 20 reviews are there. So this will be our filtering criteria

### First we will filter out the user with less than 100 reviews

```python
required = book_with_rating.groupby("User-ID").count()["Book-Title"]
required_user = required[required>=100].index
```

```python
book_with_rating
```

```python
book_with_rating = book_with_rating[book_with_rating["User-ID"].isin(required_user)]
```

```python
book_with_rating.shape
```

```python
### Here we can see that number of rows has been decrease from around 11 lac to 5 lax
```

### Lets remove the book on which less than 20 reviews are there

```python
required_data = book_with_rating.groupby("Book-Title").count()["Book-Author"]
required_book = required_data[required_data>=60].index
```

```python
book_with_rating = book_with_rating[book_with_rating["Book-Title"].isin(required_book)]
```

```python
book_with_rating.shape
```

```python
### Now we have around 82 thousand data remains
```

### Lets make a pivot table we contains book name in index and user_id in columns and book rating in values

```python
final_data_required = book_with_rating.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
```

```python
final_data_required.fillna(0,inplace=True)
```

```python
final_data_required
```

### Now our data become vector in 1639 dimensional space and the vector whose distance is close are said to be book of similar genera or taste no we will do the same with the help of cosine similarity score function from sklearn library lets see what happens

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(final_data_required)
```

```python
similarity_score
```

```python
similarity_score.shape
```

```python
def suggest_me_book(x):
    index = np.where(final_data_required.index == x)[0][0]
    similar_book = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:11]
    data = []
    for i in similar_book:
        temp = book_with_rating[book_with_rating["Book-Title"]==final_data_required.index[i[0]]]
        variable = []
        variable.extend(list(temp.drop_duplicates("Book-Title")["Book-Title"].values))
        variable.extend(list(temp.drop_duplicates("Book-Title")["Book-Author"].values))
        variable.extend(list(temp.drop_duplicates("Book-Title")["Image"].values))
        variable.extend(list(temp.drop_duplicates("Book-Title")["Publisher"].values))
        #print(final_data_required.index[i[0]])
        
        data.append(variable)
        
    return data
```
